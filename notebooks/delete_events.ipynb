```python
import requests
requests
events = requests.get("http://127.0.0.1:8000/api/events/").json()
```

```python
for event in events["results"]:
    print(f"Deleting event {event['id']}...")
    response = requests.delete(f"http://127.0.0.1:8000/api/events/{event['id']}/")
    if response.status_code == 204:
        print(f"  Event {event['id']} deleted successfully.")
    else:
        print(f"  Failed to delete event {event['id']}: HTTP {response.status_code}")
```
